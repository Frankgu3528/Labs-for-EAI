6.使用numpy创建一个10x10形状，元素在`[0,1]`均匀分布的矩阵，并且将其中小于0.5的元素置为0，其他元素不变
例如：
```python
[[0.1, 0.6],      [[0.0, 0.6],
[0.8, 1.34]]  ->  [0.8, 0.0]]
```

````python
x=np.linspace(0,1,100).reshape(10,10)
x[x<0.5]=0
print(x)
````

7.创建两个长度为100的向量，求向量的点积和夹角(提示:使用np.linalg的函数)

````python
v1=np.array([1,2])
v2=np.array([1,3])
print(np.vdot(v1,v2))
print(np.linalg.norm(v1))
print(np.linalg.norm(v2))
print(np.arccos(np.vdot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))))
````

8. 给定一个随机整数矩阵input，和一个mask整数矩阵，请你找出input和mask值相同的坐标，并且在input中把这些位置置为0

````python
import numpy as np
input = np.random.randint(1, 1000, size=(3, 1000, 1000))
mask = np.random.randint(1,1000, size=(3, 1000,1000))

# complete your code
# maybe there's a better way.
for i in range(3):
    for j in range(1000):
        for k in range(1000):
            if input[i][j][k]==mask[i][j][k]:
                input[i][j][k]=0
print(input)
````
